```python
import pandas as pd
import matplotlib.pyplot as plt

# 1080dpi
plt.rcParams['figure.dpi'] = 1080
# 
plt.rcParams['font.sans-serif'] = ['Times New Roman']  # 

# 
data = {
    'Factors': ['Slope', 'Population density', 'NPP', 'NDVI', 'LAI', 
            'Road density', 'Poi density', 'GDP', 'Night light', 'DEM'],
    'Influence': [0.408, 0.641, 0.597, 0.768, 0.634, 
              0.309, 0.221, 0.537, 0.344, 0.666]
}

# DataFrame
df = pd.DataFrame(data)
df = df.sort_values(by='Influence', ascending=False)

# 
plt.figure(figsize=(10, 6))
plt.barh(df['Factors'], df['Influence'], color='skyblue')
plt.xlabel('Influence')
# plt.title('')
plt.gca().invert_yaxis()  # y
plt.show()
```
