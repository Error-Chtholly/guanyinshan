```python
import numpy as np
import matplotlib.pyplot as plt
# 设置字体为Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
# 设置字体大小为14
plt.rcParams['font.size'] = 17
plt.rcParams['figure.dpi'] = 720

Year = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
Water_Areas = [240, 190, 190, 180, 170, 228, 433]
Trees = [79071, 75514, 75394, 75492, 75929, 79514, 75152]
Flooded_Vegetation_Areas = [251, 163, 171, 38, 3, 14, 22]
Crops = [3328, 3173, 2787, 4259, 822, 336, 559]
Built_Areas = [19704, 20971, 21196, 21533, 21820, 21236, 21360]
Bare_Groud_Areas = [317, 66, 23, 0, 22, 75, 238]
Rangelands_Areas = [11668, 14502, 14818, 13077, 15813, 13176, 16809]
Carbon_Storage = [1.514, 1.474, 1.471, 1.469, 1.47, 1.505, 1.463]

# 对Water_Areas，Trees，Flooded_Vegetation_Areas，Crops，Built_Areas，Bare_Groud_Areas，Rangelands_Areas，Carbon_Storage进行数据归一化，不要出现0和1值
# Water_Areas = (np.array(Water_Areas) - np.min(Water_Areas)) / (np.max(Water_Areas) - np.min(Water_Areas))
# Trees = (np.array(Trees) - np.min(Trees)) / (np.max(Trees) - np.min(Trees))
# Flooded_Vegetation_Areas = (np.array(Flooded_Vegetation_Areas) - np.min(Flooded_Vegetation_Areas)) / (np.max(Flooded_Vegetation_Areas) - np.min(Flooded_Vegetation_Areas))
# Crops = (np.array(Crops) - np.min(Crops)) / (np.max(Crops) - np.min(Crops))
# Built_Areas = (np.array(Built_Areas) - np.min(Built_Areas)) / (np.max(Built_Areas) - np.min(Built_Areas))
# Bare_Groud_Areas = (np.array(Bare_Groud_Areas) - np.min(Bare_Groud_Areas)) / (np.max(Bare_Groud_Areas) - np.min(Bare_Groud_Areas))
# Rangelands_Areas = (np.array(Rangelands_Areas) - np.min(Rangelands_Areas)) / (np.max(Rangelands_Areas) - np.min(Rangelands_Areas))
# Carbon_Storage = (np.array(Carbon_Storage) - np.min(Carbon_Storage)) / (np.max(Carbon_Storage) - np.min(Carbon_Storage))
# 对Water_Areas，Trees，Flooded_Vegetation_Areas，Crops，Built_Areas，Bare_Groud_Areas，Rangelands_Areas，Carbon_Storage进行数据归一化，不要出现0和1值
Water_Areas = (np.array(Water_Areas) - np.mean(Water_Areas)) / np.std(Water_Areas)
Trees = (np.array(Trees) - np.mean(Trees)) / np.std(Trees)
Flooded_Vegetation_Areas = (np.array(Flooded_Vegetation_Areas) - np.mean(Flooded_Vegetation_Areas)) / np.std(Flooded_Vegetation_Areas)
Crops = (np.array(Crops) - np.mean(Crops)) / np.std(Crops)
Built_Areas = (np.array(Built_Areas) - np.mean(Built_Areas)) / np.std(Built_Areas)
Bare_Groud_Areas = (np.array(Bare_Groud_Areas) - np.mean(Bare_Groud_Areas)) / np.std(Bare_Groud_Areas)
Rangelands_Areas = (np.array(Rangelands_Areas) - np.mean(Rangelands_Areas)) / np.std(Rangelands_Areas)
Carbon_Storage = (np.array(Carbon_Storage) - np.mean(Carbon_Storage)) / np.std(Carbon_Storage)

# 绘制数据分布图，画在一张图里面，x轴为Year，y轴为各个指标的分布情况，点线图
plt.figure(figsize=(14, 14))
plt.plot(Year, Water_Areas, label='Water Areas', color='#0070FF', marker='o', markersize=10, linewidth=3)
plt.plot(Year, Trees, label='Trees', color='#267300', marker='o', markersize=10, linewidth=3)
plt.plot(Year, Flooded_Vegetation_Areas, label='Flooded Vegetation Areas', color='#55FF00', marker='o', markersize=10, linewidth=3) # 颜色等于浅绿
plt.plot(Year, Crops, label='Crops', color='#FFFF00', marker='o', markersize=10, linewidth=3)
plt.plot(Year, Built_Areas, label='Built Areas', color='#FF0000', marker='o', markersize=10, linewidth=3)
plt.plot(Year, Bare_Groud_Areas, label='Bare Groud Areas', color='#FFAA00', marker='o', markersize=10, linewidth=3)
plt.plot(Year, Rangelands_Areas, label='Rangelands Areas', color='#DF73FF', marker='o', markersize=10, linewidth=3)
plt.plot(Year, Carbon_Storage, label='Carbon Storage', color='Black', marker='o', markersize=10, linewidth=3)
plt.xlabel('Year')
plt.ylabel('Normalized Value')
# plt.title('Distribution of Environmental Indicators')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams['font.family'] = 'Times New Roman'
# 设置字体大小为14
plt.rcParams['font.size'] = 26
# 图片分辨率
plt.rcParams['figure.dpi'] = 720

# Data
Year = [2017, 2018, 2019, 2020, 2021, 2022, 2023]
Water_Areas = [240, 190, 190, 180, 170, 228, 433]
Trees = [79071, 75514, 75394, 75492, 75929, 79514, 75152]
Flooded_Vegetation_Areas = [251, 163, 171, 38, 3, 14, 22]
Crops = [3328, 3173, 2787, 4259, 822, 336, 559]
Built_Areas = [19704, 20971, 21196, 21533, 21820, 21236, 21360]
Bare_Ground_Areas = [317, 66, 23, 0, 22, 75, 238]
Rangelands_Areas = [11668, 14502, 14818, 13077, 15813, 13176, 16809]

# Combine data into a DataFrame
data = {
    'Year': Year,
    'Water Areas': Water_Areas,
    'Trees': Trees,
    'Flooded Vegetation Areas': Flooded_Vegetation_Areas,
    'Crops': Crops,
    'Built Areas': Built_Areas,
    'Bare Ground Areas': Bare_Ground_Areas,
    'Rangelands Areas': Rangelands_Areas
}
df = pd.DataFrame(data)

# Normalize the data to make the total length of each bar the same
df_normalized = df.iloc[:, 1:].div(df.iloc[:, 1:].sum(axis=1), axis=0)

# Scale smaller categories to make them more visible
# For example, multiply smaller categories by a factor (e.g., 10)
df_normalized['Water Areas'] *= 10
df_normalized['Flooded Vegetation Areas'] *= 10
df_normalized['Crops'] *= 10
df_normalized['Bare Ground Areas'] *= 10

# Re-normalize after scaling
df_normalized = df_normalized.div(df_normalized.sum(axis=1), axis=0)

# Plot
categories = df_normalized.columns
colors = ['#0070FF', '#267300', '#55FF00', '#FFFF00', '#FF0000', '#FFAA00', '#DF73FF']

plt.figure(figsize=(21, 11))
left = np.zeros(len(Year))

for i, category in enumerate(categories):
    plt.barh(Year, df_normalized[category], left=left, label=category, color=colors[i], height=0.61) # 宽度为0.61
    left += df_normalized[category]

plt.xlabel('Proportion')
plt.ylabel('Year')
# plt.title('Horizontal Stacked Bar Chart of Land Cover Categories (2017~2023)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1.021))
plt.yticks(Year)
plt.tight_layout()
plt.show()
```
